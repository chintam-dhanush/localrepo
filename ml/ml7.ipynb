```python
heartDisease = pd.read_csv('/content/heart.csv')

print("First 5 rows of dataset:")
print(heartDisease.head())
```

```python
from pgmpy.models import DiscreteBayesianNetwork

model = DiscreteBayesianNetwork([
    ('age', 'heartdisease'),
    ('gender', 'heartdisease'),
    ('exang', 'heartdisease'),
    ('cp', 'heartdisease'),
    ('heartdisease', 'restecg'),
    ('heartdisease', 'chol')
])

print("\nBayesian Network Structure:")
print(model.edges())
```

```python
model.fit(heartDisease, estimator=MaximumLikelihoodEstimator)

print("\nConditional Probability Tables (CPTs):")
for cpd in model.get_cpds():
    print(cpd)
```

```python
print("\nInferencing with Bayesian Network:")

HeartDiseasetest_infer = VariableElimination(model)
```

```python
# Query 1
print("\n1. Probability of Heart Disease given evidence = restecg")

q1 = HeartDiseasetest_infer.query(
    variables=['heartdisease'],
    evidence={'restecg': 1}
)

print(q1)
```

```python
q2 = HeartDiseasetest_infer.query(
    variables=['heartdisease'],
    evidence={'cp': 2}
)

print(q2)
```
